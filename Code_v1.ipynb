#### Inventory Policy

```python
def inventory (col):
    c1 = col[0]
    c2 = col[1]

    if c1 == 1:
        return c2 * 0.5
    elif c1 == 2:
        return c2 * 0.75
    else:
        return c2
```

```python
df2['safety_stock'] = df2[['std_sigma','average_daily_quantity']].apply( inventory, axis =1)
```

```python
df2['total_inventory']=df2['average_daily_quantity']+df2['safety_stock']
df2['total_inventory']=df2['total_inventory'].apply(np.ceil)
```

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets=apriori(dataset_association2, min_support=0.0005,use_colnames=True) #run the association rule algorithm
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules

rules=association_rules(frequent_itemsets, metric="lift", min_threshold=0.8)
```

```python
rules["antecedent_len"] = rules["antecedents"].apply(lambda x: len(x))
rules["consequents_len"] = rules["consequents"].apply(lambda x: len(x))
df_rules=rules[ (rules['antecedent_len'] == 1) &(rules['consequents_len'] ==1) &
       (rules['conviction'] >1)
        ]
df_rules
```

```python
antecedents=[]
for i in df_rules['antecedents'].values:
    a=list(i)[0]
    antecedents.append(a)
#antecedents
```

```python
consequence=[]
for i in df_rules['consequents'].values:
    a=list(i)[:]
    consequence.append(a)
#consequence
```

```python
#Check Product Weight
def p_weight (orders_:dict,product_id_):
    if product_id_ in orders_.keys():        
        return orders_[product_id_][2]
    else:
        pass
        
```

```python
#Check Pod Max Weight
def pod_weight (dict_product_:dict,pods_layout_,first_position_i):
    pod_weight=0
    for i in range(0,pods_layout_.shape[0]):
        for j in range(0,pods_layout_.shape[1]):
            for k in range(0,len(pods_layout_[i][j])):
                if pods_layout_[first_position_i][j][k] in dict_product_.keys():
                    pod_weight+=dict_product_[pods_layout_[first_position_i][j][k]][2]                    
                else:
                    pass
        return pod_weight       
```

```python
#picking function

def picking (orders_picking_:dict,pods_layout_,product_id_:str):
    quantity_=orders_picking_[product_id_][1]
    product_picked=[]
    product_backorder=[]
    items_picked=0
    items_backorder=0
    result=False
    position_i_=[]
    
       
    while quantity_>0:
        try:
                    
            for i in range(0,pods_layout_.shape[0]):
                for j in range(0,pods_layout_.shape[1]):
                    for k in range (0,3):
                        if pods_layout_[i][j][k]== product_id_:
                            product_picked.append(product_id_)
                            pods_layout_[i][j][k]=0
                            position_i_.append(i)
                            result=True
                            if result==True:                                 
                                quantity_-=1                          
                                items_picked+=1
                                raise StopIteration

                            
                        
          

            product_backorder.append(product_id_)
            items_backorder+=1
            break
                                
                                
        
        except StopIteration:
            break
            
                    


    print('product_picked', product_picked)
    print('product_backorder',product_backorder)
    print('items_picked',items_picked,'items_backorder',items_backorder)
```
